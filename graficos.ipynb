```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data/points_tmw.csv")
```

```python
group_prod = (df.groupby("descProduto")["idTransacao"].count().reset_index())
group_prod = group_prod.sort_values(by = "idTransacao")
```

```python
sns.barplot(group_prod, y = "descProduto", x = "idTransacao")
plt.title("Quantidade de Transações")
plt.xlabel("Quantidade Transações")
plt.ylabel("Descrição Produto")
plt.show()
```

```python
df["dataTransacao"] = pd.to_datetime(df["dtTransacao"]).dt.date

group_data = df.groupby("dataTransacao").agg(
    {
        "qtdPontos": "sum",
        "idTransacao": "count",
    }
).reset_index()

group_data = group_data.sort_values(by = "dataTransacao")

plt.Figure(figsize = (10,8))
plt.plot(group_data["dataTransacao"], group_data["idTransacao"])
plt.title("Série Histórica de Transações")
plt.ylabel("Quantidade Transações")
plt.xlabel("Data de Transações")
plt.show()
```

```python
plt.hist(group_data["qtdPontos"], bins = 18)
plt.title("Frequência de pontos")
plt.xlabel("Quantidade Pontos")
plt.ylabel("Frequência")
plt.show()
```

```python
plt.boxplot(group_data["qtdPontos"])
plt.xlabel("Pontos")
plt.show()
```

```python
group_ponto = (df.groupby("qtdPontos")["idTransacao"]
               .count()
               .reset_index())

sns.barplot(group_ponto, x = "qtdPontos", y = "idTransacao")
plt.title("Transação de pontos")
plt.xlabel("Pontos")
plt.ylabel("Transações")
plt.show()
```
