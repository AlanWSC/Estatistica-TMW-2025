```python
import pandas as pd
import sqlalchemy
```

```python
df = pd.read_csv("Data/points_tmw.csv")
```

# Abrindo conexão com o banco

```python
engine = sqlalchemy.create_engine("sqlite:///data/tmw.db")
df.to_sql("points", engine, if_exists = "replace", index = False)
```

```python
df.head()
```

```python
frequencia_categoria = (df.groupby(["descCategoriaProduto"])[["idTransacao"]].count())
frequencia_categoria
```

```python
frequencia_categoria["Frequência Absoluta Acumulada"] = frequencia_categoria[["idTransacao"]].cumsum()
frequencia_categoria
```

```python
frequencia_categoria["Frequência Relativa"] = frequencia_categoria["idTransacao"] / frequencia_categoria["idTransacao"].sum()
frequencia_categoria
```

```python
frequencia_categoria["Frequência Relativa Acumulada"] = frequencia_categoria["Frequência Relativa"].cumsum()
frequencia_categoria
```
